Тема “Множество”

Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.
*Выполнить задание 1 на языке Python


```python
a={1,3,6,8,5,43,6}
b={4,7,3,2,6}
c={43,1,2}
```

```python
#объединение a и b
res = a.union(b)
res
```

```python
#пересечение a и b
res = a.intersection(b)
res
```

```python
#разность a и b
res = a.difference(b)
res
```

```python
#разность b и a
res = b.difference(a)
res
```

```python
#симметрическая разность b и a
res = a.symmetric_difference(b)
res
```

```python
# декартово произведение a и b
import itertools
for i in itertools.product(a,b):
    print(i)
```

На языке Python предложить алгоритм вычисляющий численно предел с заданной точностью 
считает до какого-то момента, потом выдает ошибку :(

```python
import math
precision = 0.0000001
n=1
while  True:
    res = n/(math.factorial(n)**(1.0/n))
    res_next = (n+1.0)/(math.factorial(n+1.0)**(1.0/(n+1.0)))
    print(res_next-res)
    
    if res_next-res < precision:
        print(res_next)
        break
    else:
         n += 1
```
